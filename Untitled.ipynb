```python
X = df.drop(['timestamp', 'label_numerical', 'Above_Threshold', 'Lightning_Strike'], axis = 1)
y = df['label_numerical']

split_idx = 4000
X_train, X_test, y_train, y_test = X[: split_idx], X[split_idx :], y[: split_idx], y[split_idx :]

print(f'Train: {X_train.shape}  SAFE={(y_train == 0).sum()}  DANGER={(y_train == 1).sum()}')
print(f'Test : {X_test.shape}   SAFE={(y_test == 0).sum()}   DANGER={(y_test == 1).sum()}')
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

print(f'Before SMOTE : SAFE={(y_train == 0).sum()}  DANGER={(y_train == 1).sum()}')
print(f'After SMOTE : SAFE={(y_train_balanced == 0).sum()}  DANGER={(y_train_balanced == 1).sum()}')
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Decision Tree'      : DecisionTreeClassifier(random_state=42),
    'Random Forest'      : RandomForestClassifier(n_estimators=200, n_jobs=-1, random_state=42),
}

cv_results = []
timeSeries_cv = TimeSeriesSplit(n_splits=5)
for name, model in models.items():
    pipe = ImbPipeline([
        ('scaler', StandardScaler()),
        ('smote',  SMOTE(random_state=42)),
        ('clf',    model)
    ])
    scores = cross_validate(pipe, X_train, y_train,cv=timeSeries_cv,
        scoring={
            'f1_macro': 'f1_macro',
                 'f1_danger': 'f1',          # f1 for positive class = DANGER
                 'roc_auc':   'roc_auc'
        }, n_jobs=-1
    )
    cv_results.append({
        'Model'         : name,
        'F1-macro'      : scores['test_f1_macro'].mean(),
        'F1-macro std'  : scores['test_f1_macro'].std(),
        'F1-DANGER'     : scores['test_f1_danger'].mean(),
        'ROC-AUC'       : scores['test_roc_auc'].mean(),
    })
    print(f"{name:<22} F1-macro={scores['test_f1_macro'].mean():.4f} "
          f"± {scores['test_f1_macro'].std():.4f}  "
          f"F1-DANGER={scores['test_f1_danger'].mean():.4f} "
         f"ROC-AUC = {scores['test_roc_auc'].mean():.4f}")

cv_df = pd.DataFrame(cv_results).sort_values('F1-macro', ascending=False)
cv_df.reset_index(drop=True).style\
     .highlight_max(subset=['F1-macro','F1-DANGER','ROC-AUC'], color='blue')\
     .highlight_min(subset=['F1-macro std'], color='blue')
```

```python
from collections import Counter

counts = Counter(y_train)
SCALE_POS_WEIGHT = counts[0] / counts[1]  # negative / positive

def xgb_objective(trial):
    params = dict(
        n_estimators      = trial.suggest_int  ('n_estimators',     200, 800),
        max_depth         = trial.suggest_int  ('max_depth',          3,   8),
        learning_rate     = trial.suggest_float('learning_rate',   0.01, 0.2, log=True),
        subsample         = trial.suggest_float('subsample',        0.6, 1.0),
        colsample_bytree  = trial.suggest_float('colsample_bytree', 0.6, 1.0),
        min_child_weight  = trial.suggest_int  ('min_child_weight',   1,  10),
        scale_pos_weight  = SCALE_POS_WEIGHT,
        tree_method       = 'hist',
        eval_metric       = 'logloss',
        random_state      = 42,
        n_jobs            = -1,
    )
    pipe = ImbPipeline([
        ('scaler', StandardScaler()),
        ('smote',  SMOTE(random_state=42)),
        ('clf',    xgb.XGBClassifier(**params)),
    ])
    return cross_val_score(pipe, X_train, y_train, cv=timeSeries_cv,
                           scoring='f1_macro', n_jobs=-1).mean()

print('Tuning XGBoost (50 trials) ...')
study_xgb = optuna.create_study(direction='maximize')
study_xgb.optimize(xgb_objective, n_trials=50, show_progress_bar=True)
print(f'  Best XGB F1-macro (CV) : {study_xgb.best_value:.4f}')
print(f'  Best params            : {study_xgb.best_params}')
```

```python
def lgb_objective(trial):
    params = dict(
        n_estimators      = trial.suggest_int  ('n_estimators',     200, 800),
        max_depth         = trial.suggest_int  ('max_depth',          3,   8),
        learning_rate     = trial.suggest_float('learning_rate',   0.01, 0.2, log=True),
        num_leaves        = trial.suggest_int  ('num_leaves',        20, 150),
        subsample         = trial.suggest_float('subsample',        0.6, 1.0),
        colsample_bytree  = trial.suggest_float('colsample_bytree', 0.6, 1.0),
        scale_pos_weight  = SCALE_POS_WEIGHT,
        random_state      = 42,
        verbose           = -1,
        n_jobs            = -1,
    )
    pipe = ImbPipeline([
        ('scaler', StandardScaler()),
        ('smote',  SMOTE(random_state = 42)),
        ('clf',    lgb.LGBMClassifier(**params)),
    ])
    return cross_val_score(pipe, X_train, y_train, cv=timeSeries_cv,
                           scoring='f1_macro', n_jobs=-1).mean()

print('Tuning LightGBM (50 trials) ...')
study_lgb = optuna.create_study(direction='maximize')
study_lgb.optimize(lgb_objective, n_trials=50, show_progress_bar=True)
print(f'  Best LGB F1-macro (CV) : {study_lgb.best_value:.4f}')
print(f'  Best params            : {study_lgb.best_params} \n')
```

```python
def tune_threshold(proba, y_true, model_name, ax):
    """Plot annotated P-R curve and return the threshold that maximises F1-DANGER."""
    precision, recall, thresholds = precision_recall_curve(y_true, proba)

    # F1-DANGER at each threshold point
    f1_vals  = 2 * precision[:-1] * recall[:-1] / (precision[:-1] + recall[:-1] + 1e-9)
    best_idx = int(f1_vals.argmax())
    best_t   = float(thresholds[best_idx])

    ax.plot(recall, precision, lw=2, color='steelblue', label='P-R curve')
    ax.scatter(recall[best_idx], precision[best_idx],
               s=120, zorder=5, color='red',
               label=f'Best threshold = {best_t:.3f}\n'
                     f'P={precision[best_idx]:.3f}  R={recall[best_idx]:.3f}')
    ax.axvline(recall[best_idx], color='red', linestyle='--', alpha=0.4)
    ax.set_xlabel('Recall  (DANGER)', fontsize=11)
    ax.set_ylabel('Precision  (DANGER)', fontsize=11)
    ax.set_title(f'{model_name} — Precision-Recall Curve', fontweight='bold')
    ax.legend(fontsize=9)
    ax.set_xlim([0, 1]); ax.set_ylim([0, 1.05])

    return best_t

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

best_t_xgb = tune_threshold(proba_xgb, y_test, 'XGBoost',  axes[0])
best_t_lgb = tune_threshold(proba_lgb, y_test, 'LightGBM', axes[1])

plt.suptitle('Threshold Tuning — Precision vs Recall (DANGER class)',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print(f'Tuned threshold → XGBoost : {best_t_xgb:.3f} | LightGBM : {best_t_lgb:.3f}')
```
